```python
from sklearn.preprocessing import StandardScaler
import numpy as np
data=np.loadtxt('/Users/abhi99/Downloads/0000000000002419_training_ccpp_x_y_train.csv',delimiter=',')
x_test=np.loadtxt('/Users/abhi99/Downloads/0000000000002419_test_ccpp_x_test.csv',delimiter=',')
x_train=data[:,:-1]

temp=np.ones((x_train.shape[0],x_train.shape[1]*4))
temp[:,:x_train.shape[1]]=x_train
temp[:,x_train.shape[1]:x_train.shape[1]*2]=x_train**2
temp[:,x_train.shape[1]*2:x_train.shape[1]*3]=x_train**3
temp[:,x_train.shape[1]*3:x_train.shape[1]*4]=x_train**4
x_train=temp

temp=np.ones((x_test.shape[0],x_test.shape[1]*4))
temp[:,:x_test.shape[1]]=x_test
temp[:,x_test.shape[1]:x_test.shape[1]*2]=x_test**2
temp[:,x_test.shape[1]*2:x_test.shape[1]*3]=x_test**3
temp[:,x_test.shape[1]*3:x_test.shape[1]*4]=x_test**4
x_test=temp


y_train=data[:,-1]
fs=StandardScaler()
x_train=fs.fit_transform(x_train)
x_test=fs.transform(x_test)

temp=np.ones((x_train.shape[0],x_train.shape[1]+1))
temp[:,:-1]=x_train
x_train=temp

temp=np.ones((x_test.shape[0],x_test.shape[1]+1))
temp[:,:-1]=x_test
x_test=temp

```

```python

def cost(x,y,m,M):
    return (1/M)*(((y-((x*m).sum(axis=1)))**2).sum())
```

```python

def stepGradient(x,y,alpha,M,N,m):
    m_slopes=np.zeros(N)
    for i in range(M):
        for j in range(N):
            m_slopes[j]+=(-2/N)*((y[i]-(m*x[i]).sum())*x[i][j])
    #print(m_slopes)
    return m-alpha*m_slopes
```

```python
def gd(x,y,alpha,itr):
    M=x.shape[0]
    N=x.shape[1]
    m=np.zeros(N)
    cst=[]
    for i in range(itr):
        m=stepGradient(x,y,alpha,M,N,m)
        cst.append(cost(x,y,m,M))
        print('Cost of iteration',i,':',cst[i])
    return m,cst
```
